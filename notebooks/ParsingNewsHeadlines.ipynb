```python
import re, unidecode
#import pandas as pd
#from io import StringIO

def matchheadlinecountrycity(line,city_country_dict,checkwithunidecode,ignoreCaseCheck):
    match = False
    city_found = ""
    countrycode = ""
    for city,country in city_country_dict.items():
        citynamefromdict = city
        if ignoreCaseCheck:
            line = line.lower()
            city = city.lower()
            if ' city' in city:
                city = city.replace(' city', '')
            if checkwithunidecode:
                city = unidecode.unidecode(city)
        #if ' city' in city:
        #    city = city.replace(' city', '')
            #citynamefromdict = citynamefromdict.replace(' City', '')
        if city in line: # Faster but does not match whole words
            regexp = re.compile(r"\b"+city+r"\b") # re.compile increases regex speed, but I did not observe any notable speed improvement, need to check further
            if regexp.search(line) and len(city) > len(city_found): # One more test with re as it matches whole words, so not to avoid city name with virus name like Rota city with Rotavirus, len(city) condition is to avoid partial match of a 2 letter city word like York matches for New York
                match = True
                city_found = citynamefromdict
                countrycode = country
    
    return (match,city_found,countrycode)

#pdlist = ["headline;city;country"]
outfile = "df_extracted.csv"
f = open (outfile,"w+")
f.write("headline;city;country\n")
for line in lines:
    city = ""
    country = countryname = ""
    (matched,city,country) = matchheadlinecountrycity(line,cities_country,checkwithunidecode=False,ignoreCaseCheck=False)
    if not matched:
        (matched,city,country) = matchheadlinecountrycity(line,cities_country,checkwithunidecode=True,ignoreCaseCheck=False)
        if not matched:
            (matched,city,country) = matchheadlinecountrycity(line,cities_country,checkwithunidecode=True,ignoreCaseCheck=True)
    if country != "":
        countryname=country_with_code[country]
    f.write(line+";"+city+";"+countryname+"\n")


f.close()
```
