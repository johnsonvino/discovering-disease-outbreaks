```python
import pandas as pd
import os

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

os.chdir("../data")

file = "df_extracted.csv"

df = pd.read_csv(file, delimiter=';')
df.head()
```

```python
import geonamescache

gc = geonamescache.GeonamesCache()

#help(gc)
country_with_code = {}
for key,values in gc.get_countries().items():
    country_with_code.__setitem__(key,values['name'])

gc.get_cities_by_name("Los Angeles")
#print(country_with_code)
```

```python
city_lat_long = {}
df['latitude'] = ""
df['longitude'] = ""
for index, row in df.iterrows():
    if not pd.isnull(row['city']): # To avoind index out of range error for NaN values99
        city = row['city']
        coordinates = [(values['latitude'],values['longitude']) for key,values in gc.get_cities_by_name(city)[0].items()]
        city_lat_long.update({city:[values['latitude'],values['longitude']] for key,values in gc.get_cities_by_name(city)[0].items()})

#print(city_lat_long)
```

```python
df['latitude'] = df['city'].map({k:v[0] for k,v in city_lat_long.items()})
df['longitude'] = df['city'].map({k:v[1] for k,v in city_lat_long.items()})
#df['country'] = df['country'].map({v:k for k,v in country_with_code.items()})
#data.dropna(inplace = True) 
headline_data = df.copy()
headline_data['country'] = headline_data['country'].map({v:k for k,v in country_with_code.items()})
nadata = df[headline_data.isnull().any(axis=1)]
nadata.headline
```
